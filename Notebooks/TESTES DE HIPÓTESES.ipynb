# **TESTES DE** **HIPÓTESES**

## **Introdução:**

Na **Geotecnia**, se trabalha com um dos materiais mais heterogéneos da engenharia: o solo. Diferente do aço ou do betão, cujas propriedades são controladas em fábrica, as propriedades do solo em uma Bacia ou em qualquer encosta são estimadas a partir de amostras limitadas.

A questão central é: *Se o valor médio de coesão que medimos num ensaio triaxial é diferente do valor de projeto, essa diferença é real ou é apenas uma flutuação estatística?*

O **Teste de Hipótese** é o procedimento estatístico que nos permite rejeitar ou não uma afirmação sobre um parâmetro populacional com base em evidências amostrais. Ele transforma a dúvida subjetiva em uma decisão fundamentada em probabilidade.

##**Conceitos Fundamentais**

**Hipótese Nula** ($H_0$): O efeito observado é apenas fruto do acaso. (*Ex: O sensor está calibrado*).

**Hipótese Alternativa** ($H_1$): Existe um efeito real ou diferença. (*Ex: O sensor apresenta erro sistemático*).

**Nível de Significância**($\alpha$): A
probabilidade de cometer um Erro Tipo I
(rejeitar $H_0$ sendo ela verdadeira). Geralmente fixado em 5%.

##**Comparação de Médias**

**Variância Conhecida (Teste Z)**

**Cenário:** *Verificação de precisão de um sensor LVDT (Linear Variable Differential Transformer)*.

**O Problema:** Um fabricante garante que o seu sensor de deslocamento tem um erro sistemático nulo ($\mu = 0$ mm) e um desvio padrão histórico conhecido de $\sigma = 0,01$ mm devido à sensibilidade do equipamento. Durante uma calibração no laboratório, você realiza 20 medições e encontra um erro médio de $0,005$ mm.

**Hipóteses**:

$H_0: \mu = 0$ (O sensor está calibrado conforme o padrão).

$H_1: \mu \neq 0$ (O sensor apresenta um desvio sistemático).

**Variância Desconhecida (Teste t-Student)**

**Cenário:** Avaliação de um novo aditivo para solo-cimento na Bacia do Recôncavo.

**O Problema:** Você desenvolveu um traço de solo-cimento e quer testar se a resistência média à compressão simples é superior a $2,1$ MPa (exigência de norma para bases de pavimentos). Como é um material novo, não conhece a variância real. Você molda 8 corpos de prova e mede as resistências.

**Hipóteses:**

$H_0: \mu = 2,1$ MPa (O material apenas cumpre o mínimo).

$H_1: \mu > 2,1$ MPa (O aditivo realmente melhora o desempenho significativamente).

##**Teste Qui-Quadrado**

Teste Qui-Quadrado ($\chi^2$) Utilizado para variáveis categóricas. Serve para testar se dois fenômenos são independentes (*Ex: O tipo de solo influencia o tipo de erosão observada?*).

**Teste Qui-Quadrado ($\chi^2$) de IndependênciaCenário:**

 ***Investigação de causas de processos erosivos.***

**O Problema:**

Imaginemos que estamos a mapear encostas e queremos saber se o tipo de vegetação (*Grama, Árvores ou Solo Exposto*) está relacionado com o tipo de erosão (*Laminar ou em Sulcos*).Recolhemos dados de 100 pontos de observação.

**Hipóteses:**

$H_0$: O tipo de erosão é independente da vegetação.

$H_1$: O tipo de erosão depende da vegetação (existe uma associação)

##**Exemplos Práticos**


**1.** Vamos imaginar que estamos testando a resistência à compressão de uma nova  istura de *solo-cimento* para estabilização de *taludes*. O valor de referência é **5.0 MPa**. Suas amostras de laboratório deram uma média ligeiramente inferior. **Vamos testar se essa diferença é significativa.**

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd

# 1. ENTRADA DE DADOS (Simulação de 10 amostras de laboratório)
# Valores ligeiramente inferiores a 5.0 para testar a significância
amostras = np.array([4.8, 4.9, 4.7, 5.1, 4.6, 4.8, 4.9, 4.5, 4.7, 5.0])
mu_referencia = 5.0
alpha = 0.05

# 2. PROCESSAMENTO ESTATÍSTICO
n = len(amostras)
media_amostral = np.mean(amostras)
desvio_padrao = np.std(amostras, ddof=1)
gl = n - 1  # Graus de liberdade

# Cálculo da estatística t (t-calculado)
t_calc = (media_amostral - mu_referencia) / (desvio_padrao / np.sqrt(n))

# Cálculo do valor crítico (t-crítico) para teste unilateral à esquerda
t_critico = stats.t.ppf(alpha, gl)

# Cálculo do P-valor
p_valor = stats.t.cdf(t_calc, gl)

# 3. CRIAÇÃO DA TABELA DE RESULTADOS
dados_tabela = {
    'Parâmetro': ['Referência (H0)', 'Média Amostral', 'Desvio Padrão (s)', 't-Calculado', 't-Crítico', 'P-valor'],
    'Valor': [mu_referencia, f"{media_amostral:.2f}", f"{desvio_padrao:.3f}", f"{t_calc:.4f}", f"{t_critico:.4f}", f"{p_valor:.4f}"]
}
df_resumo = pd.DataFrame(dados_tabela)

print("--- RELATÓRIO DE ENSAIO GEOTÉCNICO ---")
print(df_resumo.to_string(index=False))

# 4. VISUALIZAÇÃO GRÁFICA
plt.figure(figsize=(10, 6))

# Gerar curva t-Student
x = np.linspace(-5, 5, 1000)
y = stats.t.pdf(x, gl)
plt.plot(x, y, 'k-', lw=2, label=f'Distribuição t-Student (gl={gl})')

# Região de Rejeição
x_rejeicao = np.linspace(-5, t_critico, 100)
plt.fill_between(x_rejeicao, stats.t.pdf(x_rejeicao, gl), color='red', alpha=0.3, label='Região de Rejeição (H1)')

# Indicações de t-calc e t-crit
plt.axvline(t_critico, color='red', linestyle='--', label=f't-crítico ({t_critico:.2f})')
plt.axvline(t_calc, color='blue', linestyle='-', marker='o', label=f't-amostra ({t_calc:.2f})')

plt.title('Teste de Hipótese: Resistência à Compressão Solo-Cimento', fontsize=14)
plt.xlabel('Estatística t', fontsize=12)
plt.ylabel('Densidade de Probabilidade', fontsize=12)
plt.legend()
plt.grid(True, alpha=0.2)
plt.show()

# 5. CONCLUSÃO AUTOMATIZADA
if t_calc < t_critico:
    print(f"\nDECISÃO: Rejeitar H0 (P-valor = {p_valor:.4f})")
    print("CONCLUSÃO: A diferença é ESTATISTICAMENTE SIGNIFICATIVA. O material não atingiu os 5.0 MPa.")
else:
    print(f"\nDECISÃO: Não Rejeitar H0 (P-valor = {p_valor:.4f})")
    print("CONCLUSÃO: A diferença NÃO é significativa. O material pode ser considerado em conformidade.")
```

 **2.Problema**
Suponhamos que esteja actuando como engenheiro fiscal em uma obra de pavimentação. A norma técnica local exige que a base de solo-cimento tenha uma resistência média à compressão simples ($\mu$) de, no mínimo, 2,1 MPa. Para validar o lote recém-executado, Foi colectando 8 amostras aleatórias no trecho.

**Pedido** Verifique estatisticamente, com 95% de confiança ($\alpha = 0,05$), se o lote pode ser aceito ou se deve ser rejeitado por inconformidade.

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt
import pandas as pd

# 1. Dados coletados em campo (8 amostras em MPa)
# Nota: Simulamos valores próximos a 2.1 para testar a sensibilidade do teste
amostras = np.array([2.05, 2.12, 1.98, 2.08, 1.95, 2.10, 2.02, 1.90])
mu_norma = 2.1
alpha = 0.05
n = len(amostras)
gl = n - 1

# 2. Cálculos Estatísticos
media_amostral = np.mean(amostras)
desvio_padrao = np.std(amostras, ddof=1)
t_calc = (media_amostral - mu_norma) / (desvio_padrao / np.sqrt(n))

# Valor crítico para teste unilateral à esquerda
t_critico = stats.t.ppf(alpha, gl)

# 3. Exibição dos Resultados em Tabela
dados_tabela = {
    'Parâmetro': ['Norma (H0)', 'Média Obtida', 'Desvio Padrão (s)', 't-Calculado', 't-Crítico'],
    'Valor': [f"{mu_norma} MPa", f"{media_amostral:.3f} MPa", f"{desvio_padrao:.3f}", f"{t_calc:.4f}", f"{t_critico:.4f}"]
}
print("--- RELATÓRIO DE FISCALIZAÇÃO ---")
print(pd.DataFrame(dados_tabela).to_string(index=False))

# 4. Construção do Gráfico de Decisão
x = np.linspace(-4, 4, 1000)
y = stats.t.pdf(x, gl)

plt.figure(figsize=(10, 5))
plt.plot(x, y, 'k-', lw=2, label=f'Distribuição t-Student (gl={gl})')

# Região de Rejeição (Inconformidade)
x_rej = np.linspace(-4, t_critico, 100)
plt.fill_between(x_rej, stats.t.pdf(x_rej, gl), color='red', alpha=0.3, label='Região de Rejeição (Lote Reprovado)')

plt.axvline(t_critico, color='red', linestyle='--', label=f't-crítico ({t_critico:.2f})')
plt.axvline(t_calc, color='blue', linestyle='-', marker='o', label=f't-amostra ({t_calc:.2f})')

plt.title('Teste de Conformidade: Resistência Solo-Cimento')
plt.xlabel('Estatística t')
plt.ylabel('Densidade de Probabilidade')
plt.legend()
plt.grid(True, alpha=0.2)
plt.show()

# 5. Veredito Final
if t_calc < t_critico:
    print("\nDECISÃO FINAL: REJEITAR H0.")
    print("VEREDITO: O lote está em INCONFORMIDADE. A resistência é estatisticamente inferior a 2,1 MPa.")
else:
    print("\nDECISÃO FINAL: NÃO REJEITAR H0.")
    print("VEREDITO: O lote pode ser ACEITO. Não há evidência de que a resistência seja inferior ao mínimo.")
```
